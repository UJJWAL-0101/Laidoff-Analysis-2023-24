```python
def get_month():
    month_list = []
    
    col = data['Date']
    for date in col:
        parts = date.split("-")
        if len(parts) >= 2:  # Ensure that the date format is valid
            month = parts[1]
            month_list.append(month)
        else:
            print("Invalid date format:", date)
    
    return month_list

layoff_month_list = get_month()
```
